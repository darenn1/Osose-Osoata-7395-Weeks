```python
url = 'http://books.toscrape.com/'

prices = []
book_names = []
image_urls = []
ratings = []

def page_links():
    for i in range(1, 51): 
        page_links = url + f"catalogue/page-{i}.html" 
        pages = requests.get(page_links)
        bs_page = bs(pages.content)
        book_lists = bs_page.find_all('li', attrs={'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
        
        for price in book_lists:
            book_price = price.find('p', attrs={'class':'price_color'})
            prices.append(book_price.string)
            
        for book in book_lists:
            book_name = book.find('h3').find('a')['title']
            book_names.append(book_name)
            
        for rating in book_lists:
            rates = rating.find('p')['class'][1]
            ratings.append(rates)
            
        for image_link in book_lists:
            image_url = image_link.select_one('div.image_container img')
            image_url = url + image_url['src'].strip('../')
            image_urls.append(image_url)
```

```python
df = pd.DataFrame(data={'Book Name':book_names, 'Image URL':image_urls, 'Prices':prices, 'Ratings':ratings}, index=range(1,2001))
```

```python
df['Ratings'] = df['Ratings'].replace({'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5})
```
